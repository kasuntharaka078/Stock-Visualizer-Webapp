```python
import yfinance as yf
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import mplfinance as mpf
from stockstats import StockDataFrame
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
```

```python
#Importing Data from Yahoo Finance API
symbol = 'BTC-USD'
period='90d'
ticker = yf.Ticker(symbol)
df = ticker.history(period)
df = df.reset_index()
df.head()
```

```python
fig = px.line(df, x=df.Date, y=["Open", "High", "Low", "Close"],
              labels={"variable": "Price Type"},
              title="Stock Prices Over Time")
fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'], high=df['High'],
                low=df['Low'], close=df['Close'])
                     ])

fig.update_xaxes(rangeslider_visible=False)

fig.update_xaxes(title_text="Date")  # Update the x-axis label
fig.update_yaxes(title_text="Value")  # Update the y-axis label

# Add a title to the chart
fig.update_layout(title="Stock Candlestick Chart")

fig.show()
```

```python
# Calculate the 10-day SMA using Pandas
df['SMA_10'] = df['Close'].rolling(window=10).mean()

# Create a Plotly Express line plot
fig = px.line(df, x=df.Date, y=['SMA_10', 'Close'], labels={'variable': 'Price Type'}, title="Stock 10-day SMA vs. Close Prices")
fig.show()
```

```python
# Calculate the 10-day SMA and EMA using Pandas
df['SMA_10'] = df['Close'].rolling(window=10).mean()
df['EMA_10'] = df['Close'].ewm(span=10, adjust=False).mean()

# Create a Plotly Express line plot
fig = px.line(df, x=df.Date, y=['SMA_10', 'EMA_10', 'Close'], labels={'variable': 'Price Type'}, title="Stock 10-day SMA and EMA vs. Close Prices")
fig.show()
```

```python
import plotly.express as px

# Calculate MACD and Signal Line using Pandas
df['12-day EMA'] = df['Close'].ewm(span=12, adjust=False).mean()
df['26-day EMA'] = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = df['12-day EMA'] - df['26-day EMA']
df['Signal Line'] = df['MACD'].ewm(span=9, adjust=False).mean()

# Create a Plotly Express line plot
fig = px.line(df, x=df.Date, y=['MACD', 'Signal Line'], labels={'variable': 'Indicator'}, title="Stock MACD and Signal Line")
fig.show()
```
